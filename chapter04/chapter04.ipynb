> 案例4-2 谁是小偷？

警察抓了A、B、C、D四个偷窃嫌疑犯，其中只有一个人是真正的小偷，审问记录如下：

+ A说：“我不是小偷。”
+ B说：“C是小偷。”
+ C说：“小偷肯定是D。”
+ D说：“C在冤枉人。”

```python
suspects = ["A", "B", "C", "D"]
for x in suspects:
    if (x != "A") + (x == "C") + (x == "D") + (x != "D") == 3:
        print("\n小偷是:", x)
        break
```

案例4-4<br>
请用筛选法求出300以内的素数。

```python
primes = [1] * 300
primes[0:2] = [0, 0]
for i in range(2, 300):
    if primes[i] == 1:
        for j in range(i + 1, 300):
            if primes[j] != 0 and j % i == 0:
                primes[j] = 0

print("300以内的素数包括：")
for i in range(2, 300):
    if primes[i]:
        print(i, end=" ")
```

案例4-5<br>
请用二分查找法在列表中查找指定的元素。

```python
ls = [34, 64, 67, 72, 73, 82, 83, 85, 87, 88, 90, 91, 96, 98]
x = int(input("请输入待查找的数:"))
low = 0
high = len(ls) - 1
```

```python
while low < high:
    mid = (low + high) // 2
    if ls[mid] < x:
        low = mid + 1
    elif ls[mid] > x:
        high = mid - 1
    else:
        print("找到{},索引为{}!".format(x, mid))
        break
else:
    print("没有找到{}".format(x))
```

**案例4-7**<br>
拟掷两个骰子100000次,统计2～12各点数出现的概率。

```python
from random import *
```

```python
faces = [0] * 13
for i in range(100000):
    face1 = randint(1, 6)
    face2 = randint(1, 6)
    faces[face1 + face2] += 1
```

```python
print("模拟掷两个骰子100000次结果如下:")
for i in range(2, 13):
    rate = faces[i] / 100000
    print("点数{}共出现了{}次".format(i, faces[i]), end=",")
    print("出现概率{:.2%}".format(rate))
```

案例4-9<br>

大侠属性评分统计。

```python
attrs = ["筋骨", "敏捷", "气势", "反应", "技巧", "内力"]
tables = [
    ["萧峰", 20, 17, 20, 20, 18, 19],
    ["杨过", 18, 19, 17, 20, 18, 18],
    ["令狐冲", 12, 17, 14, 20, 19, 13],
    ["张无忌", 20, 17, 15, 14, 20, 20],
    ["郭靖", 19, 18, 19, 18, 19, 20],
]
```

大侠名字列表names

```python
names = [item[0] for item in tables]
```

评分列表scores

```python
scores = [item[1:] for item in tables]
```

各位大侠的总分列表totals

```python
totals = [sum(item) for item in scores]
```

各个属性平均分列表avgs

```python
avgs = []
for j in range(6):
    avgs.append(sum([scores[i][j] for i in range(5)]) / 5)
```

五位大侠的总分

```python
print("\n五位大侠的总分是:")
for i in range(5):
    print("{:<6}:{:>4}".format(names[i], totals[i]))
```

不同属性的平均分

```python
print("\n不同属性的平均分是:")
for i in range(6):
    print("{:<8}:{:>4}".format(attrs[i], avgs[i]))
```

得分最高的大侠名字

```python
print("\n总分最高的大侠是:", end="——")
print(names[totals.index(max(totals))])
```
